```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


file_path =  r"C:\Users\85355\Downloads\11_final_data.csv"
df = pd.read_csv(file_path)

sustaining_color = "green"
non_sustaining_color = "orange"

df["time_taken_to_merge"] = pd.to_numeric(df["time_taken_to_merge"], errors="coerce")
df["ratio_of_compatibility_issues"] = pd.to_numeric(df["ratio_of_compatibility_issues"], errors="coerce")

df["contributed_back_forks_count"] = pd.to_numeric(df["contributed_back_forks_count"], errors="coerce")
df["hard_forks_count"] = pd.to_numeric(df["hard_forks_count"], errors="coerce")
df["merged_commits_count"] = pd.to_numeric(df["merged_commits_count"], errors="coerce")
df["not_merged_commits_count"] = pd.to_numeric(df["not_merged_commits_count"], errors="coerce")
df["time_taken_to_merge"] = pd.to_numeric(df["time_taken_to_merge"], errors="coerce")

plt.figure(figsize=(10, 6))
sns.histplot(df[df["is_sustaining"] == 1]["project_age"], bins=20, kde=True, color=sustaining_color, label="Sustaining", alpha=0.6)
sns.histplot(df[df["is_sustaining"] == 0]["project_age"], bins=20, kde=True, color=non_sustaining_color, label="Non-Sustaining", alpha=0.6)
plt.title("Distribution of Project Age (Sustaining vs. Non-Sustaining)")
plt.xlabel("Project Age (Years)")
plt.ylabel("Frequency")
plt.legend()
plt.show()

```
